* Class

This is a numerical variable that indicates which class was each passenger travelling. In the Titanic specific case, it is an important feature since it is known that physical segregation of classes (with locked doors and barriers) were a normal feature in boats of this era. Since the hit occurred during the night, these physical barriers might have played a role in allowing women and children from 2nd and 3rd class to reach the lifeboats in time. Indeed, there were more survivors in first class (over 60%) than in second (just under 50%) and third (around 25%). 

```python
df_version_one = df_lean_two.copy()
df_version_one['Survived'].groupby(df_version_one['Pclass']).mean().plot.bar()
```

Class does not need any changes so far and is ready to enter the model.

```python
age_class = []
age_title = [[], [], []]
title = df_lean_two['title'].unique()
title.sort()
pclass = df_lean_two['Pclass'].unique()
pclass.sort()

c = 1

while c <= 3:
    for a in title:
        cond = ((df_lean_two['title'] == a) & (df_lean_two['Pclass'] == c))
        y = 'Age'
        new_age_mean = df_lean_two.loc[cond, y].isnull().sum()
        age_class.append(new_age_mean)
        age_title[c-1].append(new_age_mean)
    c = c+1

null_age = pd.DataFrame(age_title, index = pclass, columns = title)
null_age.T
```

There are several missing values in different classes and titles. The most common were on Mr travelling in 3rd class, then Miss also travelling in 3rd class. By knowing the median age of each class, we can impute that value for each of the missing values. 

```python
df_age_imputation = df_lean_two.copy()
df_age_imputation.head(20)
```

```python
median_age = []
c = 1
title = df_age_imputation['title'].unique()
title.sort()
pclass = df_age_imputation['Pclass'].unique()
pclass.sort()

df_age_imputation['Age'].fillna(-5, inplace = True)

while c <= 3:
    for a in title:
        cond = ((df_lean_two['title'] == a) & (df_lean_two['Pclass'] == c))
        y = 'Age'
        median_age = df_lean_two.loc[cond, y].median()
        cond_2 = ((df_age_imputation['title'] == a) & (df_age_imputation['Pclass'] == c) & (df_age_imputation['Age'] == -5))
        df_age_imputation.loc[cond_2, 'Age'] = median_age
        #print(median_age)
        median_age = []
        
    c = c+1
```

This variable indicated the ticket number for each passenger and there were no missing values. There is a lot of variability in this column though, which took careful attention and some online research to understand what it all meant. Most tickets were composed by 6 digits, where the first digit indicates class. Some were 7-digit numbers, where the first digit also indicated class. 4-digit tickets tended to correspond to passengers travelling in 2nd and 3rd class (but there were two exceptions). Most 5-digit tickets belonged to passengers traveling either in 1
st or 2nd class, though there were 18 exceptions of tickets travelling in 3rd class, with 11 of them having board Titanic in Queenstown). Tickets 
beginning with A belong to people that travelling in 3rd class and embarked in Southampton (with one exception that embarked in Queenstown). Tickets beginning with C belonged to people that either travelled in 2nd or 3rd class and embarked in Southampton. Tickets that stated with F belong to people traveling in 2nd class (two exceptions, one in 1st and another in 3rd) and boarded in Southampton. LINE tickets were of people working in the boat. Tickets starting with P belong mainly to passengers travelling in 1st class (with two exceptions travelling in 2nd three exceptions travelling in 3rd) and boarded in Cherbourg (some exceptions that boarded in Southampton). Tickets that started with S belonged to passengers travelling either in 2nd or 3rd class and boarded in Southampton (13 exceptions that boarded in Cherbourg). Finally, tickets starting with W belonged to passengers of all classes and embarked in Southampton.

```python
ticket_number = df_version_three['Ticket'].value_counts()
dict_ticket = ticket_number.to_dict()
dict_ticket
```

```python
df_version_three['ticket_repeated'] = df_version_three['Ticket'].map(dict_ticket)
df_version_three.head()
```

```python
df_version_three['travels_with_SibSp'] = df_version_three['SibSp'].map(lambda x: 1 if x > 0 else 0)
df_version_three.head()
```

```python
df_version_three['travels_with_Parch'] = df_version_three['Parch'].map(lambda x: 1 if x > 0 else 0)
df_version_three.head()
```

```python
df_version_three['travels_family'] = df_version_three['travels_with_SibSp']+df_version_three['travels_with_Parch']
df_version_three.head()
```

```python
# let's binarize ticket_repeated - 0 if it is not repeated, 1 if it is repeated (no matter the number of times)
df_version_three['ticket_binary'] = df_version_three['ticket_repeated'].map(lambda x: 1 if x > 1 else 0)
df_version_three.head()
```

```python
# Let's binarize family_travels - 0 if is not traveling with any parents or siblings, 1 if it is
df_version_three['travels_family'] = df_version_three['travels_family'].map(lambda x: 1 if x > 0 else 0)
df_version_three['travels_family']
```

```python
# Let's create a travel_accompanied column by adding travels_family to ticket_binary 
# 0 if ticket does not repeat nor travels with family - probably travels alone, 
# 1 if ticket repeats or travels with family - probably travels in group/family
# 2 if ticket repeats and travels with family - certainly travels in family

df_version_three['travels_accompanied'] = df_version_three['ticket_binary'] + df_version_three['travels_family']
df_version_three.head()
```

```python
df_version_four = df_version_three.drop(['Ticket', 'Embarked', 'last_name', 'title', 'name', 'travels_with_SibSp', 'travels_with_Parch', 
                                         'travels_family', 'ticket_binary'], axis = 1)
```
